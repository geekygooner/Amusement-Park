```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Processed.csv')
```

```python
df=df.drop(df.columns[0], axis=1)
```

```python
df_summer=df[df['Season']=='Summer']
df_fall=df[df['Season']=='Fall']
df_spring=df[df['Season']=='Spring']
df_winter=df[df['Season']=='Winter']
```

```python
sns.pairplot(x_vars=['Humidity'],y_vars=['Ticket1'],data=df,hue='Season',size=5);
plt.title('Humidity vs Ticket1');
```

```python
sns.pairplot(x_vars=['StandardTemperature'],y_vars=['Ticket1'],data=df,hue='Season',size=5);
plt.title('StandardTemperature vs Ticket1');
```

```python
sns.pairplot(x_vars=['Wind'],y_vars=['Ticket1'],data=df,hue='Season',size=5);
plt.title('Wind vs Ticket1');
```

```python
sns.pairplot(x_vars=['Humidity'],y_vars=['Ticket2'],data=df,hue='Season',size=5);
plt.title('Humidity vs Ticket2');
```

```python
sns.pairplot(x_vars=['StandardTemperature'],y_vars=['Ticket2'],data=df,hue='Season',size=5);
plt.title('StandardTemperature vs Ticket2');
```

```python
sns.pairplot(x_vars=['Wind'],y_vars=['Ticket2'],data=df,hue='Season',size=5);
plt.title('Wind vs Ticket2');
```

```python
df=df[['StandardTemperature', 'Humidity', 'Wind','Day','Weekend', 'Season','Ticket1', 'Ticket2']]
df_dummies=pd.get_dummies(df['Season'])
df_new=pd.concat([df,df_dummies],axis=1)
df_new=df_new[['StandardTemperature', 'Humidity', 'Wind','Weekend','Fall', 'Spring', 'Summer', 'Winter','Ticket1', 'Ticket2']]
corr=df_new.corr()
corr
```

```python
sns.heatmap(corr,cmap="YlGnBu",linewidths=0.5,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values);
plt.title('Correlation Heatmap');
```

```python
df_new.to_csv('Preprocessed_v2.csv',index=False)
```
